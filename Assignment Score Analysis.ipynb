# what next

- We can not use final Exam score for further analysis, it sould be removed from dataset
- For course GGG, we need to use non-weighted assignment score for further analysis
    - Approach: use average score of each student 
- For course BBB(2014J) and FFF has both graded and non-graded assessment 
    - we can drop or ignore non-graded assessment score for these courses
    
# prepare the datasets

- remove final exam score 
- remove non-weighted assessmnet of course BBB(2014J) and course FFF :: not necessary --> weight*score=0
- For course GGG, calculate the average score of each student
- For rest of the course, sum the assessment score based on (weight * score / 100)

```python
# remove final exam score
print('Before remove: ', studAssessment.shape)
studAssessment = studAssessment[~studAssessment['id_assessment'].isin(exam_id)]
print('After remove: ', studAssessment.shape)
```

```python
# merge the assessmnet and studAssessment table
merge_assessment = pd.merge(assessment,studAssessment, on=['id_assessment'], how='inner')
merge_assessment
```

```python
# multiply score and weight for all courses except, GGG
merge_assessment['final_marks'] = merge_assessment.apply(lambda x: x[5]*x[9]/100 if x[0]!='GGG' else x[9], axis=1)
merge_assessment
```

```python
final_assessment = pd.concat([merge_assessment.query('code_module!="GGG"').
                        groupby(['code_module', 'code_presentation', 'id_student'])['final_marks'].sum(),
                        merge_assessment.query('code_module=="GGG"').
                        groupby(['code_module', 'code_presentation', 'id_student'])['final_marks'].mean()
                             ])
final_assessment
```

```python
final_assessment = final_assessment.reset_index()
final_assessment['final_marks'] = final_assessment['final_marks'].map(lambda x: round(x,2))
final_assessment
```

# Lets merge final assessmnet data with studentInfo table

```python
col_list = ['code_module', 'code_presentation', 'id_student','final_result']
studentinfo = pd.read_csv('./data/studentInfo.csv', usecols=col_list)
studentinfo
```

```python
# merge the table
student_assessment = pd.merge(studentinfo, final_assessment, 
                              on=['code_module', 'code_presentation', 'id_student'],
                             how='left')
student_assessment
```
