```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sns.set_style('whitegrid')
```

```python
# Read the StudentInfo file
studinfo = pd.read_csv('./data/studentInfo.csv')
studinfo
```

```python
studvle = pd.read_csv('./data/studentVle.csv')
studvle
```

```python
group_result = studvle.groupby(['code_module', 'code_presentation', 'id_student', 'date'])['sum_click'].sum()
group_result
```

```python
dd = pd.DataFrame(np.zeros((len(studinfo), 295)), columns=range(-25, 270))
final = pd.concat([studinfo[['code_module', 'code_presentation', 'id_student', 'final_result']], dd], axis=1)
final
```

```python
i=0
for row in final.iterrows():
    m = row[1]['code_module']  # row[1][0] throw error 
    t = row[1]['code_presentation']
    s = row[1]['id_student']
    try:
        ss = group_result.loc[(m, t, s, ), ].T
        cols = list(ss.index)
        vals = list(ss.values)
        final.loc[row[0], cols] = vals
    except:
        i+=1
print('No record match: ', i)
```

```python
final.to_csv('stud_dates_click.csv', index=False)
```
