```python
mega['marks'] = mega['weight'] * mega['score']/100
mega
```

```python
group = mega.groupby(['code_module', 'code_presentation', 'id_student']).sum()
group
```

```python
group.reset_index(inplace=True)
group.query('weight>100')['code_module'].unique()
```

```python
group.to_csv('stud_graded_nonGraded_assessment.csv')
```

```python
query = 'date<{} and weight!=0'.format(270)
weighted = group.query(query)
weighted
```

```python
weighted = weighted.groupby(['code_module', 'code_presentation', 'id_student'])['marks'].sum()
weighted
```

```python
query = 'date<{} and weight==0'.format(100)
non_weighted = group.query(query)
non_weighted = non_weighted.groupby(['code_module', 'code_presentation', 'id_student'])['marks'].mean()
non_weighted
```

```python
assessment_score = pd.merge(weighted, non_weighted, on=['code_module', 'code_presentation', 'id_student'], how='outer')
```

```python
assessment_score.columns = ['graded', 'non_graded']
assessment_score.fillna(0, inplace=True)
assessment_score
```
