```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
studAssessment = pd.read_csv('./data/studentAssessment.csv')
studAssessment
```

```python
assessment = pd.read_csv('./data/assessments.csv')
assessment
```

```python
# find out number of assessmnet associated with each course
assessment.groupby(['code_module', 'code_presentation'])['id_assessment'].count()
```

```python
# find all the assessments which weights is 0
weight0 = list(assessment.query('weight==0')['id_assessment'].values)

# find student score on assignmnet which weights is 0 to verify whether students
# takes 0 weights assignmnet seriouslu or nor
studAssessment[studAssessment['id_assessment'].apply(lambda x: x in weight0)]
```

```python
studAssessment[studAssessment['id_assessment'].apply(lambda x: x in weight0)].describe()
```

### The above result shows that majority of students takes 0 weighted assignmnets seriously 

```python
# marge the assessment table with studentassessment
mega = pd.merge(assessment, studAssessment, on=['id_assessment'], how='inner')
mega
```

```python
group = mega.groupby(['code_module', 'code_presentation', 'id_student', 'id_assessment'])[['date', 'date_submitted', 'weight', 'score']].mean()
group
```
