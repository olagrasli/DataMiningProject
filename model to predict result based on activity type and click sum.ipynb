```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./student_activityType_sumClick.csv')
data
```

```python
studinfo = pd.read_csv('./data/studentInfo.csv')
studinfo = studinfo[['code_module', 'code_presentation', 'id_student', 'final_result']]
studinfo
```

```python
merge = pd.merge(studinfo, data, on=['code_module', 'code_presentation', 'id_student'], how='inner')
merge
```

# drop those records who has no records in studentVLE table

```python
merge = merge[~(merge.iloc[:, 4:].sum(axis=1)==0)]
merge
```

```python
courses = list(merge['code_module'].unique())
times = list(merge['code_presentation'].unique())
course_map = {course:i for i, course in enumerate(courses, start=1)}
times_map = {time:i for i, time in enumerate(times, start=1)}
print(course_map, times_map)
```

```python
merge['final_result'] = merge['final_result'].map({'Pass':1, 'Distinction':2, 'Fail':0, 'Withdrawn':0})
merge['code_module'] = merge['code_module'].map(course_map)
merge['code_presentation'] = merge['code_presentation'].map(times_map)

X = merge.drop(columns=['id_student', 'final_result'])
y = merge['final_result']
```

```python
X.shape, y.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
```

```python
# Random forest classifier
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
```

```python
# predict
predict = rfc.predict(X_test)
print('#### confusion metrix ####')
print(confusion_matrix(predict, y_test))
print('####### Classification report #######')
print(classification_report(predict, y_test))
```

# Support Vector Classifier

```python
svc = SVC()
svc.fit(X_train, y_train)
```

```python
# predict
predict = svc.predict(X_test)
print('#### confusion metrix ####')
print(confusion_matrix(predict, y_test))
print('####### Classification report #######')
print(classification_report(predict, y_test))
```

# Gradient Boost Classifier

```python
gbc = GradientBoostingClassifier()
gbc.fit(X_train, y_train)
```

```python
# predict
predict = gbc.predict(X_test)
print('#### confusion metrix ####')
print(confusion_matrix(predict, y_test))
print('####### Classification report #######')
print(classification_report(predict, y_test))
```

# Deep Learning Approach using Tensorflow

```python
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras import Model, Sequential
from tensorflow.keras.utils import to_categorical
```

```python
model = Sequential()
model.add(Input(shape=(X_train.shape[1], )))
model.add(Dense(32, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(128, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, to_categorical(y_train), validation_split=0.2, epochs=20, batch_size=32, verbose=2)
```

```python
predict = model.predict(X_test)
predict = np.argmax(predict, axis=1)
print('#### confusion metrix ####')
print(confusion_matrix(predict, y_test))
print('####### Classification report #######')
print(classification_report(predict, y_test))
```
