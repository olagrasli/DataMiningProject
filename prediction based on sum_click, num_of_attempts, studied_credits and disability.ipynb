```python
path = "C:/Users/polit/OneDrive/Desktop/Jypyter notebook/OULAD/dataset/studentInfo.csv"
studentInfo = pd.read_csv(path)
studentInfo
```

```python
path = "C:/Users/polit/OneDrive/Desktop/Jypyter notebook/OULAD/dataset/studentVle.csv"
studentVle = pd.read_csv(path)
studentVle
```

```python
demo_click = pd.merge(studentVle, studentInfo, on = ['id_student','code_module','code_presentation'], how = 'left')
demo_click
```

```python
demo_click.head()
```

```python
data = demo_click.replace("Pass", "1").replace("Distinction", "1").replace("Withdrawn", "0").replace("Fail", "0").replace("N",0).replace("Y",1)
data 
```

```python
# Convert dataframe to numpy arrary
data=data.to_numpy()
```

```python
data.shape
```

```python
training = data[0:9000000]
testing = data[9000000:]
```

```python
print(training.shape)
print(testing.shape)
```

```python
training_features = training[:,[5,11,12,13]]
training_labels = training[:,-1]
testing_features = testing[:,[5,11,12,13]]
testing_labels = testing[:,-1]
```

```python
print(training_features.shape)
print(training_labels.shape)
```

```python
training_features = np.asarray(training_features).astype(np.float32)
training_labels = np.asarray(training_labels).astype(np.float32)
testing_features = np.asarray(testing_features).astype(np.float32)
testing_labels = np.asarray(testing_labels).astype(np.float32)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout,BatchNormalization
```

```python
# Model definition
model = Sequential()
model.add(Input(shape=(4,)))
model.add(Dense(16,activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.2))
model.add(Dense(8,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='softmax'))
```

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(training_features,training_labels,epochs=3,validation_data=(testing_features,testing_labels))
```
