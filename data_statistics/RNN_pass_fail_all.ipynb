## Format data to training and test set

```python
# day and week boundaries
min_startday = -25
max_endday = 270
weeks = int((max_endday-min_startday)/7)

# resource activities
activities = list(set(list(merged_vle_df['activity_type'])))
```

```python
def get_click_matrix(student_id, course_vle_df): 
    oneStudentVle = course_vle_df.loc[student_id, :]
    click_matrix = np.zeros((len(activities), weeks))

    # insert clicks
    for i, vle_row in oneStudentVle.iterrows(): 
        # extract matrix indecies
        week_index = int((vle_row['date']-min_startday-1)/7)
        resource_index = activities.index(vle_row['activity_type'])
        # add clicks to right cell
        click_matrix[resource_index, week_index] += vle_row['sum_click']
        
    return click_matrix

def get_pass_fail(student_id, course_studentInfo_df): 
    result_encoded = int(course_studentInfo_df.loc[course_studentInfo_df['id_student']==student_id, 'result_encoded'])
    
    return result_encoded
```

```python
# format the data
x, y = [], []
# iterate through all courses
for index, row in courses_df.iterrows(): 
    code_module = row['code_module']
    code_presentation = row['code_presentation']
    print(code_module, code_presentation)
    
    # extract only vle stats from the given course
    course_vle_df = merged_vle_df[(merged_vle_df.code_module==code_module)&(merged_vle_df.code_presentation==code_presentation)]

    # set index to id_student for easier extraction based on indivisual students later
    course_vle_df = course_vle_df.set_index('id_student')

    # extract student information for course module
    course_studentInfo_df = studentInfo_df.loc[(studentInfo_df.code_module==code_module)&(studentInfo_df.code_presentation==code_presentation), :]
    
    # format x and y data for model training 
    student_ids = list(set(course_vle_df.index))
    
    for student_id in student_ids: 
        try: 
            click_matrix = get_click_matrix(student_id, course_vle_df)
            result_encoded = get_pass_fail(student_id, course_studentInfo_df)
            x.append(click_matrix)
            y.append(result_encoded)
        except Exception as e: 
            pass
            #print(student_id)
            print(e)
# numpy format
X = np.array(x)
y = np.array(y)
```

## Split data into training and test set

```python
from torch import nn
import torch
import torch.optim as optim
import torch.nn.functional as F
```

```python
# check if gpu is available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 3)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state = 3)
```

```python
# make datasets into torch tensors
X_train, y_train = torch.from_numpy(X_train).float(), torch.tensor(y_train).float()
X_test, y_test = torch.from_numpy(X_test).float(), torch.tensor(y_test).float()
X_val, y_val = torch.from_numpy(X_val).float(), torch.tensor(y_val).float()
```

```python
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)
X_val = X_val.to(device)
y_val = y_val.to(device)
```

## Create dataloader

```python
from torch.utils.data import DataLoader, Dataset
```

```python
class oversampdata(Dataset):
    def __init__(self, x_data, targets):
        self.data = x_data
        self.targets = targets

    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, index):
        target = self.targets[index]
        data_val = self.data[index]
        
        return data_val, target
```

```python
train_dataset = oversampdata(X_train, y_train)
valid_dataset = oversampdata(X_test, y_test)
```

```python
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader = DataLoader(valid_dataset, batch_size=64, shuffle=False)
```

## Make and train pytorch model

```python
class RNN(nn.Module): 
    def __init__(self, input_size, hidden_size, output_size, n_layers=1): 
        super(RNN, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.n_layers = n_layers
        
        #self.encoder = nn.Embedding(input_size, hidden_size)
        self.lstm = nn.LSTM(input_size, hidden_size, n_layers, bidirectional = True)
        self.fc1 = nn.Linear(hidden_size*2*20, 16)
        self.fc2 = nn.Linear(16, output_size)
        
    def forward(self, inp): 
        #e = self.encoder(inp)
        output, hidden = self.lstm(inp)
        #print(output.shape)
        #print(output.view(-1, output.size(0)).shape)
        
        #x = self.fc1(output[:, -1, :])
        x = self.fc1(output.view( output.size(0), -1) )
        x = F.relu(x)
        x = self.fc2(x)
        x = torch.sigmoid(x)
        #x = torch.nn.Softmax(dim=1)(x)
        
        return x
```

```python
# define model instance
rnn_net = RNN(X_train[0].shape[1], 8, 1, n_layers = 1)
rnn_net_best_val = RNN(X_train[0].shape[1], 8, 1, n_layers = 1)

# set loss function, optimizer and criterion
criterion = nn.BCELoss()
optimizer = optim.Adam(rnn_net.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
```

```python
# set training to gpu if exists
rnn_net.to(device)
rnn_net_best_val.to(device)

# training parameters
n_steps = 30
best_val_loss = 100
```

```python
# train model
for i in range(n_steps):
    for i, data in enumerate(train_loader, 0):        
        inputs, labels = data        
        optimizer.zero_grad()
        y_pred_train = rnn_net(inputs)
        loss_train = criterion(y_pred_train,labels.unsqueeze(1))        
        loss_train.backward()        
        optimizer.step()
        
    # validation loss
    y_pred_val = rnn_net(X_val)
    loss_val = criterion(y_pred_val, y_val.unsqueeze(1))
    if (best_val_loss > loss_val.cpu().detach().numpy()): 
        rnn_net_best_val.load_state_dict(rnn_net.state_dict())
        best_val_loss = loss_val.cpu().detach().numpy()
        print("Updated best val model")

    # print loss
    print(loss_val.cpu().detach().numpy())
```

```python
def binary_acc(y_pred, y_test):
    y_pred = y_pred.cpu()
    y_test = y_test.cpu()
    y_pred_tag = torch.round(y_pred.squeeze(1))

    correct_results_sum = (y_pred_tag == y_test).sum().float()
    acc = correct_results_sum/y_test.shape[0]
    #acc = torch.round(acc * 100)
    
    return acc
```

```python
y_eval = rnn_net_best_val(X_test)
binary_acc(y_eval, y_test)
```

```python
y_pred_binary = torch.round(y_eval.cpu().squeeze(1))
```

```python
print(classification_report(y_test.cpu().detach().numpy(), y_pred_binary.detach().numpy()))
```

```python
# remove models from gpu
rnn_net.to('cpu')
rnn_net_best_val.to('cpu')
```

## Dense model

```python
class DenseNN(nn.Module): 
    def __init__(self, input_size, output_size): 
        super(DenseNN, self).__init__()
        self.fc1 = nn.Linear(input_size, 32)
        self.fc2 = nn.Linear(32, 16)
        self.fc3 = nn.Linear(16, output_size)
        
    def forward(self, inp): 
        # propogate through densely connected linear layers
        x = inp.view(inp.shape[0],-1)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.fc2(x)
        x = F.relu(x)
        x = self.fc3(x)
        x = torch.sigmoid(x)
        
        return x
```

```python
# define model instance
input_size = X_train.shape[1]*X_train.shape[2]
dense_net = DenseNN(input_size, 1)
dense_net_best_val = DenseNN(input_size, 1)

# set loss function, optimizer and criterion
criterion = nn.BCELoss()
optimizer = optim.Adam(dense_net.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
```

```python
# set training to gpu if exists
dense_net.to(device)
dense_net_best_val.to(device)

# training parameters
n_steps = 20
best_val_loss = 100
```

```python
# train model
for i in range(n_steps):
    for i, data in enumerate(train_loader, 0):        
        inputs, labels = data        
        optimizer.zero_grad()
        y_pred_train = dense_net(inputs)
        loss_train = criterion(y_pred_train,labels.unsqueeze(1))        
        loss_train.backward()        
        optimizer.step()
        
    # validation loss
    y_pred_val = dense_net(X_val)
    loss_val = criterion(y_pred_val, y_val.unsqueeze(1))
    if (best_val_loss > loss_val.cpu().detach().numpy()): 
        dense_net_best_val.load_state_dict(dense_net.state_dict())
        best_val_loss = loss_val.cpu().detach().numpy()
        print("Updated best val model")

    # print loss
    print(loss_val.cpu().detach().numpy())
```

## CNN model

```python
class CNN(nn.Module):    
    def __init__(self):        
        super(CNN,self).__init__()      
        self.conv1 = nn.Conv2d(1,6,kernel_size = 5)        
        self.pool = nn.MaxPool2d(2,2)        
        self.conv2 = nn.Conv2d(6,16,5)        
        self.fc1 = nn.Linear(224,120)        
        self.fc2 = nn.Linear(120,84)        
        self.fc3 = nn.Linear(84,1)    
        
    def forward(self,x):   
        x = x.view(x.shape[0], 1, x.shape[1], x.shape[2])
        x = self.pool(F.relu(self.conv1(x)))        
        x = self.pool(F.relu(self.conv2(x)))        
        x = x.view(x.size(0), -1)        
        x = F.relu(self.fc1(x))        
        x = F.relu(self.fc2(x))        
        x = F.sigmoid(self.fc3(x))        

        return x
```

```python
# define model instance
cnn_net = CNN()
cnn_net_best_val = CNN()

# set loss function, optimizer and criterion
#criterion = nn.CrossEntropyLoss()
#criterion = nn.BCEWithLogitsLoss()
criterion = nn.BCELoss()
optimizer = optim.Adam(cnn_net.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
```

```python
# set training to gpu if exists
cnn_net.to(device)
cnn_net_best_val.to(device)


# training parameters
n_steps = 50
best_val_loss = 100
```

```python
# train model
for i in range(n_steps):
    for i, data in enumerate(train_loader, 0):        
        inputs, labels = data        
        optimizer.zero_grad()
        y_pred_train = cnn_net(inputs)
        loss_train = criterion(y_pred_train,labels.unsqueeze(1))        
        loss_train.backward()        
        optimizer.step()
        
    # validation loss
    y_pred_val = cnn_net(X_val)
    loss_val = criterion(y_pred_val, y_val.unsqueeze(1))
    if (best_val_loss > loss_val.cpu().detach().numpy()): 
        cnn_net_best_val.load_state_dict(cnn_net.state_dict())
        best_val_loss = loss_val.cpu().detach().numpy()
        print("Updated best val model")

    # print loss
    print(loss_val.cpu().detach().numpy())
```
