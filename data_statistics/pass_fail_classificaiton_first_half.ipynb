## Format data to training and test set

```python
# extrect activity types
activities = list(set(course_vle_df['activity_type']))

# extract day-information
days = sorted(list(set(course_vle_df['date'])))
first_day, last_day = days[0], days[-1]
duration = last_day-first_day
weeks = int(duration/7)+1
```

```python
def get_click_matrix(student_id, course_vle_df=course_vle_df, weeks=weeks, first_day=first_day, activities=activities): 
    oneStudentVle = course_vle_df.loc[student_id, :]
    click_matrix = np.zeros((weeks, len(activities)))

    # insert clicks
    for i, vle_row in oneStudentVle.iterrows(): 
        # extract matrix indecies
        week_index = int((vle_row['date']-first_day)/7)
        resource_index = activities.index(vle_row['activity_type'])
        # add clicks to right cell
        click_matrix[week_index, resource_index] += vle_row['sum_click']
        
    # only include the first half of the activity statistics
    half_index = int(weeks/2)
    half_click_matrix = click_matrix[half_index:, :]
        
    return half_click_matrix

def get_pass_fail(student_id, course_studentInfo_df=course_studentInfo_df): 
    one_hot_result = np.zeros(3)
    result_encoded = int(course_studentInfo_df.loc[course_studentInfo_df['id_student']==student_id, 'result_encoded'])
    #one_hot_result[result_encoded] = 1
    
    return result_encoded
```

```python
# format x and y data for model training 
student_ids = list(set(course_vle_df.index))
x, y = [], []
for student_id in student_ids: 
    try: 
        click_matrix = get_click_matrix(student_id)
        result_encoded = get_pass_fail(student_id)
        x.append(np.transpose(click_matrix))
        y.append(result_encoded)
    except Exception as e: 
        print(student_id)
        print(e)
X = np.array(x)
y = np.array(y)
```

## Split data into training and test set

```python
from torch import nn
import torch
import torch.optim as optim
import torch.nn.functional as F
```

```python
# check if gpu is available
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 3)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state = 3)
```

```python
# make datasets into torch tensors
X_train, y_train = torch.from_numpy(X_train).float(), torch.tensor(y_train).float()
X_test, y_test = torch.from_numpy(X_test).float(), torch.tensor(y_test).float()
X_val, y_val = torch.from_numpy(X_val).float(), torch.tensor(y_val).float()
```

## Make and train pytorch model

```python
class RNN(nn.Module): 
    def __init__(self, input_size, hidden_size, output_size, n_layers=1): 
        super(RNN, self).__init__()
        self.input_size = input_size
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.n_layers = n_layers
        
        #self.encoder = nn.Embedding(input_size, hidden_size)
        self.lstm = nn.LSTM(input_size, hidden_size, n_layers, bidirectional = True)
        self.fc1 = nn.Linear(hidden_size*2, 32)
        self.fc2 = nn.Linear(32, output_size)
        
    def forward(self, inp): 
        #e = self.encoder(inp)
        output, hidden = self.lstm(inp)
        
        x = self.fc1(output[:, -1, :])
        x = F.relu(x)
        x = self.fc2(x)
        x = torch.sigmoid(x)
        #x = torch.nn.Softmax(dim=1)(x)
        
        return x
```

```python
# define model instance
rnn_net = RNN(X_train[0].shape[1], 16, 1, n_layers = 2) #y_train[0].shape[0]
rnn_net_best_val = RNN(X_train[0].shape[1], 16, 1, n_layers = 2)

# set loss function, optimizer and criterion
#criterion = nn.CrossEntropyLoss()
#criterion = nn.BCEWithLogitsLoss()
criterion = nn.BCELoss()
optimizer = optim.Adam(rnn_net.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
```

```python
# set training to gpu if exists
rnn_net.to(device)
rnn_net_best_val.to(device)
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)
X_val = X_val.to(device)
y_val = y_val.to(device)

# training parameters
n_steps = 50
best_val_loss = 100
```

```python
# train model
for i in range(n_steps):
    # training loss
    y_pred_train = rnn_net(X_train)
    loss_train = criterion(y_pred_train, y_train.unsqueeze(1))
    # validation loss
    y_pred_val = rnn_net(X_val)
    loss_val = criterion(y_pred_val, y_val.unsqueeze(1))
    if (best_val_loss > loss_val.cpu().detach().numpy()): 
        rnn_net_best_val.load_state_dict(rnn_net.state_dict())
        best_val_loss = loss_val.cpu().detach().numpy()
        print("Updated best val model")
    
    # run optimzaiton algorithm
    optimizer.zero_grad()
    loss_train.backward()
    optimizer.step()
    # print loss
    print(loss_train.cpu().detach().numpy())
```

```python
def binary_acc(y_pred, y_test):
    y_pred = y_pred.cpu()
    y_test = y_test.cpu()
    y_pred_tag = torch.round(y_pred.squeeze(1))

    correct_results_sum = (y_pred_tag == y_test).sum().float()
    acc = correct_results_sum/y_test.shape[0]
    acc = torch.round(acc * 100)
    
    return acc
```

```python
y_eval = rnn_net_best_val(X_test)
binary_acc(y_eval, y_test)
```

```python
y_pred_binary = torch.round(y_eval.cpu().squeeze(1))
```

```python
print(classification_report(y_test.cpu().detach().numpy(), y_pred_binary.detach().numpy()))
```

## Include demographics

```python
# get demographics info from dataframe
demographics_df = course_studentInfo_df[['id_student', 'gender', 'region', 'highest_education', 'imd_band', 'age_band']]
```

```python
one_hot_demo_df = pd.get_dummies(demographics_df, columns=['gender', 'region', 'highest_education', 'imd_band', 'age_band'])
```

```python
def get_demographics(student_id, one_hot_demo_df=one_hot_demo_df): 
    one_hot_demographics = one_hot_demo_df[one_hot_demo_df['id_student']==student_id].drop(columns=['id_student']).values[0]
    
    return one_hot_demographics
```

```python
# format x, demographics and y data for model training 
student_ids = list(set(course_vle_df.index))
x, demographics, y = [], [], []
for student_id in student_ids: 
    try: 
        click_matrix = get_click_matrix(student_id)
        result_encoded = get_pass_fail(student_id)
        one_hot_demographics = get_demographics(student_id)
        x.append(np.transpose(click_matrix))
        demographics.append(one_hot_demographics)
        y.append(result_encoded)
    except Exception as e: 
        print(student_id)
        print(e)
X = np.array(x)
demographics = np.array(demographics)
y = np.array(y)
```

```python
X_train, X_test, demo_train, demo_test, y_train, y_test = train_test_split(X, demographics, y, test_size=0.2, random_state = 3)
X_train, X_val, demo_train, demo_val, y_train, y_val = train_test_split(X_train, demo_train, y_train, test_size=0.1, random_state = 3)
```

```python
# make datasets into torch tensors
X_train, demo_train, y_train = torch.from_numpy(X_train).float(), torch.tensor(demo_train).float(), torch.tensor(y_train).float()
X_test, demo_test, y_test = torch.from_numpy(X_test).float(), torch.tensor(demo_test).float(), torch.tensor(y_test).float()
X_val, demo_val, y_val = torch.from_numpy(X_val).float(), torch.tensor(demo_val).float(), torch.tensor(y_val).float()
```

## Make and train pytorch model with demographics

```python
class RNN_2(nn.Module): 
    def __init__(self, input_size, input_size2, hidden_size, output_size, n_layers=1): 
        super(RNN_2, self).__init__()
        self.input_size = input_size
        self.input_size2 = input_size2
        self.hidden_size = hidden_size
        self.output_size = output_size
        self.n_layers = n_layers
        
        # feed forward dense neural networks
        self.fc_demo = nn.Linear(input_size2, 16)
        
        # recurrent neural networks for vle clicks
        self.lstm = nn.LSTM(input_size, hidden_size, n_layers, bidirectional = True)
        self.fc_vle = nn.Linear(hidden_size*2, 32)
        
        # last output layer
        self.fc_out = nn.Linear(32+16, output_size)
        
    def forward(self, inp_vle, inp_demo): 
        
        # forward pass vle part
        output_vle, hidden = self.lstm(inp_vle)
        x_vle = self.fc_vle(output_vle[:, -1, :])
        x_vle = F.relu(x_vle)
        
        # forward pass demographics
        x_demo = self.fc_demo(inp_demo)
        x_demo = F.relu(x_demo)
        
        # concatenate data from both input layers
        x = torch.cat((x_vle, x_demo), dim=1)
        
        # last output layer
        x = self.fc_out(x)
        x = torch.sigmoid(x)
        
        return x
```

```python
# define model instance
rnn_net = RNN_2(X_train[0].shape[1], demo_train[0].shape[0], 16, 1, n_layers = 2)
rnn_net_best_val = RNN_2(X_train[0].shape[1], demo_train[0].shape[0], 16, 1, n_layers = 2)

# set loss function, optimizer and criterion
criterion = nn.BCELoss()
optimizer = optim.Adam(rnn_net.parameters(), lr=0.01)
loss_fn = nn.MSELoss()
```

```python
# set training to gpu if exists
rnn_net.to(device)
rnn_net_best_val.to(device)
X_train = X_train.to(device)
y_train = y_train.to(device)
demo_train = demo_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)
demo_test = demo_test.to(device)
X_val = X_val.to(device)
y_val = y_val.to(device)
demo_val = demo_val.to(device)

# training parameters
n_steps = 50
best_val_loss = 100
```

```python
# train model
for i in range(n_steps):
    # training loss
    y_pred_train = rnn_net(X_train, demo_train)
    loss_train = criterion(y_pred_train, y_train.unsqueeze(1))
    
    # validation loss
    y_pred_val = rnn_net(X_val, demo_val)
    loss_val = criterion(y_pred_val, y_val.unsqueeze(1))
    if (best_val_loss > loss_val.cpu().detach().numpy()): 
        rnn_net_best_val.load_state_dict(rnn_net.state_dict())
        best_val_loss = loss_val.cpu().detach().numpy()
        print("Updated best val model")
    
    # optimize model 
    optimizer.zero_grad()
    loss_train.backward()
    optimizer.step()
    # print loss
    print(loss_train.cpu().detach().numpy())
```
