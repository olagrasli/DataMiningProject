```python
studentInfo_df['result_encoded'] = studentInfo_df['final_result'].map({'Pass':1, 'Distinction':1, 'Withdrawn':0, 'Fail':0})
```

## Merge pandas frame to get data-format

```python
# first join studentVLE with vle to get full view of vle activity
merged_vle_df = pd.merge(studentVle_df, vle_df, on = ['id_site', 'code_module', 'code_presentation'], how='left')
```

```python
merged_vle_df
```

## Format data to training and test set

```python
# day and week boundaries
min_startday = -25
max_endday = 270
weeks = int((max_endday-min_startday)/7)

# resource activities
activities = list(set(list(merged_vle_df['activity_type'])))
```

```python
def get_click_matrix(student_id, course_vle_df): 
    oneStudentVle = course_vle_df.loc[student_id, :]
    click_matrix = np.zeros((len(activities), weeks))

    # insert clicks
    for i, vle_row in oneStudentVle.iterrows(): 
        # extract matrix indecies
        week_index = int((vle_row['date']-min_startday-1)/7)
        resource_index = activities.index(vle_row['activity_type'])
        # add clicks to right cell
        click_matrix[resource_index, week_index] += vle_row['sum_click']
        
    return click_matrix

def get_pass_fail(student_id, course_studentInfo_df): 
    result_encoded = int(course_studentInfo_df.loc[course_studentInfo_df['id_student']==student_id, 'result_encoded'])
    final_result = str(course_studentInfo_df.loc[course_studentInfo_df['id_student']==student_id, 'final_result'].item())
    
    return result_encoded, final_result
```

```python
# format the data
x, y, course_modules, results = [], [], [], []
# iterate through all courses
for index, row in courses_df.iterrows(): 
    code_module = row['code_module']
    code_presentation = row['code_presentation']
    print(code_module, code_presentation)
    
    # extract only vle stats from the given course
    course_vle_df = merged_vle_df[(merged_vle_df.code_module==code_module)&(merged_vle_df.code_presentation==code_presentation)]

    # set index to id_student for easier extraction based on indivisual students later
    course_vle_df = course_vle_df.set_index('id_student')

    # extract student information for course module
    course_studentInfo_df = studentInfo_df.loc[(studentInfo_df.code_module==code_module)&(studentInfo_df.code_presentation==code_presentation), :]
    
    # format x and y data for model training 
    student_ids = list(set(course_vle_df.index))
    
    for student_id in student_ids: 
        try: 
            click_matrix = get_click_matrix(student_id, course_vle_df)
            result_encoded, final_result = get_pass_fail(student_id, course_studentInfo_df)
            x.append(click_matrix)
            y.append(result_encoded)
            course_modules.append((code_module, code_presentation))
            results.append(final_result)
        except Exception as e: 
            pass
            #print(student_id)
            print(e)
# numpy format
X = np.array(x)
y = np.array(y)
```

## Save formated data into DataFrame

```python
# create dataframe from X, Y, course_modules and results
result_format_df = pd.DataFrame({'X': list(X), 'y': list(y), 'course_modules': course_modules, 'results': results})
```

# Scatterplots to visualize student activity

```python
import random
import torch
```

```python
# pass / fail students
X[y==1].shape, X[y==0].shape
```

```python
passes = [result == 'Pass' for result in results]
random_indicies_passes = random.sample(range(X[passes].shape[0]), 20)
random_indicies_passes = torch.tensor(random_indicies_passes)

distinctions = [result == 'Distinction' for result in results]
random_indicies_distinct = random.sample(range(X[distinctions].shape[0]), 20)
random_indicies_distinct = torch.tensor(random_indicies_distinct)

fails = [result == 'Fail' for result in results]
random_indicies_fails = random.sample(range(X[fails].shape[0]), 20)
random_indicies_fails = torch.tensor(random_indicies_fails)

withdrawns = [result == 'Withdrawn' for result in results]
random_indicies_withdrawn = random.sample(range(X[withdrawns].shape[0]), 20)
random_indicies_withdrawn = torch.tensor(random_indicies_withdrawn)
```

```python
random_passes = X[passes][random_indicies_passes]
random_distincts = X[distinctions][random_indicies_distinct]
random_fails = X[fails][random_indicies_fails]
random_withdrawn = X[withdrawns][random_indicies_withdrawn]
```

```python
# plot activity for pass students
fig, axes = plt.subplots(4, 5, figsize=(13,10))
fig.suptitle("Click Activity Pass Students", fontsize=20)
for j, random_pass in enumerate(random_passes): 
    # plot
    sub_row = int(j/5)
    sub_col = j % 5
    axes[sub_row, sub_col].axis([0, 42, 0, 150])

    for resource_clicks in random_pass: 
        weeks, click_weeks = [], []
        # extract only the weeks with actual activity
        for i, resource_click in enumerate(resource_clicks): 
            if(resource_click != 0): 
                weeks.append(i)
                click_weeks.append(resource_click)
        axes[sub_row, sub_col].scatter(weeks, click_weeks, c='b', s=20, alpha=0.3)
        
# set labels
plt.setp(axes[-1, :], xlabel='Week')
plt.setp(axes[:, 0], ylabel='Weekly Clicks')

plt.show()
```

## Plot class distribution for whole dataset

```python
cnt_final_results = Counter(list(studentInfo_df['final_result']))

class_labels = list(cnt_final_results.keys())
class_counts = list(cnt_final_results.values())
```

```python
class_labels, class_counts
```

```python
# Pie chart, where the slices will be ordered and plotted counter-clockwise:
fig1, ax1 = plt.subplots()
ax1.pie(class_counts, labels=class_labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# find the distribution of binary data
binary_labels = ['Pass/ Distinctions', 'Fail/ Withdrawn']
binary_counts = [class_counts[0]+class_counts[3], class_counts[1]+class_counts[2]]
```
