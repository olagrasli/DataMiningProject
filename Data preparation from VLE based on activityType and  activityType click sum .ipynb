```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

# studentVle

```python
studvle = pd.read_csv('./data/studentVle.csv')
studvle
```

# vle

```python
vle = pd.read_csv('./data/vle.csv')
vle
```

```python
site_activity = {}
for ids, ac in zip(vle['id_site'], vle['activity_type']):
    site_activity[ids] = ac
```

```python
# map the site id by actiity_type
studvle['id_site'] = studvle['id_site'].map(site_activity)
studvle
```

```python
group_result = studvle.groupby(['code_module', 'code_presentation', 'id_student', 'id_site'])['id_site', 'sum_click'].agg({'id_site':'count', 'sum_click':'sum'})
group_result
```

# lets create columns based on activity type and click sum 

```python
cols_open = []
cols_click = []
for name in vle['activity_type'].value_counts().index:
    cols_open.append(name)
    cols_click.append(name+'_click')
```

```python
all_cols = cols_open+cols_click
all_cols
```

```python
studinfo = pd.read_csv('./data/studentInfo.csv')

dd = pd.DataFrame(np.zeros((len(studinfo), 40)), columns=all_cols)
newdf = pd.concat([studinfo[['code_module', 'code_presentation','id_student']], dd], axis=1)
newdf
```

```python
i=0
for row in newdf.iterrows():
    m = row[1][0]
    t = row[1][1]
    s = row[1][2]
    try:
        ss = group_result.loc[(m, t, s, ), ].T
        cols = list(ss.loc['id_site'].index)
        vals = list(ss.loc['id_site'].values)
        newdf.loc[row[0], cols] = vals
        cols = [col+'_click' for col in cols]
        vals = list(ss.loc['sum_click'].values)
        newdf.loc[row[0], cols] = vals
    except:
        i+=1
print('No record match: ', i)
```

```python
newdf.to_csv('student_activityType_sumClick.csv', index=False)
```
